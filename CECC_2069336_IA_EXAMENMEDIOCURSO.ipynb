**Parte 1: Carga y Limpieza (Pandas y NumPy)**

1. Importa las librerías necesarias (pandas, numpy, matplotlib.pyplot).

```python
# Importar las librerías necesarias
# Importamos la librería pandas para manipulación y análisis de datos en estructuras tipo DataFrame
import pandas as pd
# Importamos numpy para operaciones matemáticas y estadísticas
import numpy as np
# Importamos matplotlib.pyplot para crear gráficas y visualizaciones
import matplotlib.pyplot as plt
```

2. Lee el archivo CSV desde la URL proporcionada y guárdalo en un DataFrame.

```python
# Guardamos en la variable url el enlace donde se encuentra el dataset en formato CSV
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
# Leemos el archivo CSV desde la URL utilizando pandas y lo guardamos en un DataFrame llamado df
df = pd.read_csv(url)
```

3. Imprime las primeras 5 filas y usa un comando para mostrar los tipos de datos de cada columna.

```python
# Imprimimos un mensaje en pantalla para identificar lo que se mostrará
print("Primeras 5 filas del dataset:")
# Mostramos las primeras cinco filas del DataFrame para observar la estructura de los datos
display(df.head())
# Imprimimos un mensaje indicando que se mostrarán los tipos de datos
print("Tipos de datos de cada columna:")
# Mostramos el tipo de dato de cada columna del DataFrame (int, float, object, etc.)
print(df.dtypes)
```

4. Identifica si hay valores nulos en la columna horsepower (caballos de fuerza).

```python
# Mostramos la cantidad de valores nulos existentes en la columna horsepower
print("Valores nulos en antes de la limpieza en la columna horsepower:")
# isnull() identifica valores nulos y sum() cuenta cuántos hay en cada columna
print(df['horsepower'].isnull().sum())
# Convertimos la columna horsepower a tipo numérico
# errors='coerce' convierte valores inválidos en NaN (valores nulos)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
# Calculamos la mediana de la columna horsepower ignorando valores nulos usando numpy
mediana_hp = np.nanmedian(df['horsepower'])
# Reemplazamos los valores nulos de la columna horsepower con la mediana calculada
df['horsepower'] = df['horsepower'].fillna(mediana_hp)
# Verificamos nuevamente si aún existen valores nulos en la columna
print("Valores nulos después de la limpieza en la columna horsepower:")
# Mostramos la cantidad de valores nulos restantes
print(df['horsepower'].isnull().sum())
```

5. Crea una nueva columna llamada relacion_peso_potencia que sea el resultado de dividir la columna weight (peso) entre horsepower.

```python
# Creamos una nueva columna llamada relacion_peso_potencia
# Esta columna divide el peso del vehículo (weight) entre los caballos de fuerza (horsepower)
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
# Mostramos las primeras filas del DataFrame para confirmar que la nueva columna se creó correctamente
print("Primeras filas con la nueva columna:")
display(df.head())
```

**Parte 2: Visualización de Datos (Matplotlib)**

 Genera las siguientes gráficas. Todas deben incluir título y etiquetas en los ejes X y Y.

1. **Histograma:** Muestra la distribución de la aceleración (acceleration) de los vehículos.

```python
# Creamos un histograma usando los valores de la columna acceleration
# bins=15 divide los datos en 15 intervalos para visualizar mejor la distribución
plt.hist(df['acceleration'], bins=15)
# Agregamos un título a la gráfica
plt.title("Distribución de la aceleración de los vehículos")
# Etiqueta del eje X
plt.xlabel("Aceleración")
# Etiqueta del eje Y
plt.ylabel("Frecuencia")
# Mostramos la gráfica generada
plt.show()
```

2. **Gráfico de Dispersión (Scatter):** Grafica el Desplazamiento del motor (displacement) en el eje X versus los Caballos de fuerza (horsepower) en el eje Y.

```python
# Creamos un gráfico de dispersión
# En el eje X colocamos displacement (desplazamiento del motor)
# En el eje Y colocamos horsepower (caballos de fuerza)
plt.scatter(df['displacement'], df['horsepower'])
# Título de la gráfica
plt.title("Relación entre desplazamiento del motor y caballos de fuerza")
# Etiqueta del eje X
plt.xlabel("Displacement (Desplazamiento del motor)")
# Etiqueta del eje Y
plt.ylabel("Horsepower (Caballos de fuerza)")
# Mostramos la gráfica
plt.show()
```

3. **Gráfico de Barras:** Muestra el promedio de rendimiento (mpg - millas por galón) agrupado por el país de origen (origin). Pista: Tendrás que agrupar los datos primero con Pandas.

```python
# Agrupamos los datos por la columna origin (país de origen)
# y calculamos el promedio del rendimiento mpg para cada grupo
mpg_promedio = df.groupby('origin')['mpg'].mean()
# Generamos un gráfico de barras a partir de los datos agrupados
mpg_promedio.plot(kind='bar')
# Agregamos título a la gráfica
plt.title("Promedio de rendimiento (MPG) por país de origen")
# Etiqueta del eje X
plt.xlabel("País de origen")
# Etiqueta del eje Y
plt.ylabel("Promedio MPG")
# Mostramos la gráfica
plt.show()
```
